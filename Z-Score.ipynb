```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set()
```

The Z-Score tells us how common or exceptional a specific observation is, represented by the number of standard deviations from removed the mean.

#### The Z-Score is defined as:
$$Z = \frac{x - \bar{x}}{\sigma}$$

##### where $\sigma = $ Standard deviation

```python
x = np.random.normal(0,0.4,100)
x = np.append(x,4)

plt.figure(figsize=(12,5))
plt.plot(kind='hist',bins=15)
plt.hist(x,bins=100)
plt.show()
```

We can see that there lies an outlier in the distribution.

```python
z_score = (x - x.mean()) / x.std()
print(f'Z-Score for the the data point that lies at (x = 4): {z_score[-1]}')
```

The image above shows the Z-Scores for a normal distribution. If the Z-Score for a particular data point. The following shows how common an occurence is:

$$-1 \leq z \leq 0 \text{ or } 0 \leq z \leq 1 \rightarrow \text{ Common}$$\
$$-2 \leq z \leq -1 \text{ or } 1 \leq z \leq 2 \rightarrow \text{ Not that common}$$\
$$-3 \leq z \leq -2 \text{ or } 2 \leq z \leq 3 \rightarrow \text{ Rare}$$\
$$-3 \leq z \leq -4 \text{ or } 3 \leq z \leq 4 \rightarrow \text{ Very rare}$$

As the Z-Score for the outlier is approximately 7.042, we can conclude that the probability for a data point to lie at $x = 4$ is quite rare.

```python
x_delta = 0.5
z_score = (x_delta - x.mean()) / x.std()
print(f'Z-Score for the the data point that lies at (x = 0.5): {z_score}')
```

As the Z-Score for a data point that lies at $x=0.5$ is approximately 0.86, we can conclude that the probability for a data point to lie at $x = 0.5$ is very high.
